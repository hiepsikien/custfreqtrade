```python
import pandas as pd
import json
```

```python
file = "../backtest_results/backtest-result-2023-03-14_23-44-45.json"
backtest_result = json.load(open(file))
daily_profit = backtest_result["strategy"]["ArbitrageBothSides"]["daily_profit"]
df = pd.DataFrame(data=daily_profit,columns=["date","profit"])
df["cumsum_profit"]=df["profit"].cumsum()
df["cumsum_profit"].plot()
```
